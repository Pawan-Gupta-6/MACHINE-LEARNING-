```python
import cv2

# Load the pre-trained face classifier
face_cascade = cv2.CascadeClassifier("C:/Users/hp/anaconda3/Lib/site-packages/cv2/data/haarcascade_frontalface_default.xml")

# Start capturing video from the default camera (0)
video_cap = cv2.VideoCapture(0)

while True:
    # Read a frame from the video feed
    ret, video_data = video_cap.read()

    # Convert the frame to grayscale
    gray = cv2.cvtColor(video_data, cv2.COLOR_BGR2GRAY)

    # Detect faces in the frame
    faces = face_cascade.detectMultiScale(
        gray,
        scaleFactor=1.1,
        minNeighbors=5,
        minSize=(30, 30),
        flags=cv2.CASCADE_SCALE_IMAGE
    )

    # Draw rectangles around the detected faces
    for (x, y, w, h) in faces:
        cv2.rectangle(video_data, (x, y), (x+w, y+h), (0, 255, 0), 2)

    # Display the resulting frame
    cv2.imshow("video_live", video_data)

    # Break the loop if 'a' key is pressed
    if cv2.waitKey(10) == ord("a"):
        break

# Release the video capture object and close all windows
video_cap.release()
cv2.destroyAllWindows()
```
